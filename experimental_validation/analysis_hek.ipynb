```python
#Modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data
filename = 'flow_cytometry_data.xlsx'
mfi_df = pd.read_excel(filename, sheet_name='MFI', header=0, usecols=list(np.arange(0,9))) 
std_fi_df = pd.read_excel(filename, sheet_name='MFI', header=0, usecols=list(np.arange(9,18))) 
std_fi_df.columns = mfi_df.columns
print(mfi_df)

mie_df = pd.read_excel(filename, sheet_name='IE', header=0, usecols=list(np.arange(0,9))) 
std_ie_df = pd.read_excel(filename, sheet_name='IE', header=0, usecols=list(np.arange(9,18))) 
std_ie_df.columns = mie_df.columns

mfi_p_df = pd.read_excel(filename, sheet_name='MFIP', header=0, usecols=list(np.arange(0,9))) 
std_fi_p_df = pd.read_excel(filename, sheet_name='MFIP', header=0, usecols=list(np.arange(9,18))) 
std_fi_p_df.columns = mfi_p_df.columns
```

```python
# plot each line separately
for i in range(mfi_df.shape[0]):
    plt.errorbar(list(mfi_df.columns)[1:], list(mfi_df.iloc[i])[1:], yerr=list(std_fi_df.iloc[i])[1:], label=mfi_df.iloc[i][0])

# add labels and legend
plt.ylabel('Mean fluorescence intensity (MFI)')

# show the plot
plt.legend(bbox_to_anchor=(1.35, 1.15), ncol = 8)
plt.yscale("log")  
plt.show()
#plt.savefig('MFI_HEK293.png', dpi=400, bbox_inches='tight')
```

```python
# plot each line separately
for i in range(mie_df.shape[0]):
    plt.errorbar(list(mie_df.columns)[1:], list(mie_df.iloc[i])[1:], yerr=list(std_ie_df.iloc[i])[1:], label=mie_df.iloc[i][0])

# add labels and legend
plt.ylabel('Stability (% EGFP-positive cells)')

# show the plot
plt.legend(bbox_to_anchor=(1.00, 1.00))
plt.show()
#plt.savefig('Stability_HEK293.png', dpi=400, bbox_inches='tight')
```

```python
# plot each line separately
for i in range(mfi_p_df.shape[0]):
    plt.errorbar(list(mfi_p_df.columns)[1:], list(mfi_p_df.iloc[i])[1:], yerr=list(std_fi_p_df.iloc[i])[1:], label=mfi_p_df.iloc[i][0])

# add labels and legend
plt.ylabel('MFI (EGFP-positive population)')

# show the plot
plt.legend(bbox_to_anchor=(1.00, 1.00))
plt.yscale("log")  
plt.show()
#plt.savefig('MFI_P_HEK293.png', dpi=400, bbox_inches='tight')
```
